```python
import pandas as pd

df = pd.read_csv('downloads/Unemployment rate in Washington - Sheet1.csv')

df.head()
```

```python
y = list(df['employment-population ratio'])
x = list(df['Period'])
z = list(df['Year'])

import matplotlib.pyplot as plt

plt.scatter(x,y,c=z, cmap='magma')
plt.xlabel("Period")
plt.ylabel("Employment-Population Ratio")

print("2019=black","2020=red","2021=yellow")
```
